```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

# Objective:
Group patients into clusters of **Hight risk of Heart attack, Medium risk and Low risk** based on similarities in their health attributes (like blood pressure, cholesterol, age, etc.) without using any labels (i.e., unsupervised learning).

# Getting Familiar with the data:

```python
df = pd.read_csv("heart_disease.csv")
df
```

```python
df.info()
```

```python
df.describe()
```

# Preprocessing The Data:

```python
df.drop(columns=['id'], inplace=True)
```

```python
df['trestbps'] = df['trestbps'].fillna(df['trestbps'].median())
df['chol'] = df['chol'].fillna(df['chol'].median())
df['thalch'] = df['thalch'].fillna(df['thalch'].median())
df['oldpeak'] = df['oldpeak'].fillna(df['oldpeak'].median())
df['ca'] = df['ca'].fillna(df['ca'].median())
```

```python
for col in ['fbs', 'restecg', 'exang', 'slope', 'thal']:
    df[col] = df[col].fillna(df[col].mode()[0])
```

# Dropping Unnecessary columns:

```python
df.drop(columns=['dataset'], inplace=True)
```

```python
df.info()
```

# One Hot Encoding Categorical Features:

```python
categorical_cols = ['sex', 'cp', 'restecg', 'slope', 'thal']
```

```python
df_encoded = pd.get_dummies(df, columns=categorical_cols, drop_first=True)
```

```python
# Identify one-hot encoded columns 
one_hot_cols = [col for col in df_encoded.columns if any(prefix + '_' in col for prefix in categorical_cols)]

# Convert only one-hot encoded columns to int
df_encoded[one_hot_cols] = df_encoded[one_hot_cols].astype(int)
```

```python
df_encoded['fbs'] = df_encoded['fbs'].astype(int)
df_encoded['exang'] = df_encoded['exang'].astype(int)
```

```python
df_encoded.head(3)
```

```python
df_encoded.info()
```

# Applying Feature Scaling:

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_data = scaler.fit_transform(df_encoded)
```

```python
# BEFORE SCALING
print("Ranges BEFORE Scaling:\n")
for col in df_encoded.columns:
    col_min = df_encoded[col].min()
    col_max = df_encoded[col].max()
    print(f"{col:30s} | min: {col_min:7.2f}, max: {col_max:7.2f}")
```

```python
scaled_df = pd.DataFrame(scaled_data, columns=df_encoded.columns)

# AFTER SCALING
print("\nRanges AFTER Scaling:\n")
for col in scaled_df.columns:
    col_min = scaled_df[col].min()
    col_max = scaled_df[col].max()
    print(f"{col:30s} | min: {col_min:7.2f}, max: {col_max:7.2f}")
```

# Model Training:

# Elbow method to choose optimal number of Clusters:

```python
from sklearn.cluster import KMeans

inertia = []
k_range = range(1, 11)

for k in k_range:
    model = KMeans(n_clusters=k, random_state=42)
    model.fit(scaled_data)
    inertia.append(model.inertia_)

plt.figure(figsize=(8,5))
plt.plot(k_range, inertia, marker='o')
plt.title('Elbow Method for Optimal k')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Inertia (Within-Cluster Sum of Squares)')
plt.grid(True)
plt.show()
```

# Applying K-means with 3 clusters:

```python
kmeans = KMeans(n_clusters=3, random_state=42)
cluster_labels = kmeans.fit_predict(scaled_data)

df_encoded['risk_cluster'] = cluster_labels
```

```python
df_encoded.head()
```

# Analyzing Each Cluster:

```python
cluster_profile = df_encoded.groupby('risk_cluster').mean().round(2)
print(cluster_profile)
```

# Mapping the Risk Labels:

```python
risk_map = {
    0: 'Low Risk',
    1: 'High Risk',
    2: 'Medium Risk'
}

df_encoded['risk_label'] = df_encoded['risk_cluster'].map(risk_map)
```

# PCA Based Cluster Visualization:

```python
from sklearn.decomposition import PCA

# Reduce to 2 dimensions with PCA
pca = PCA(n_components=2)
pca_data = pca.fit_transform(scaled_data)

# Add PCA components and cluster labels to a new DataFrame
viz_df = pd.DataFrame(pca_data, columns=['PCA1', 'PCA2'])
viz_df['Cluster'] = df_encoded['risk_label']

# Plot the clusters
plt.figure(figsize=(10, 7))
sns.scatterplot(data=viz_df, x='PCA1', y='PCA2', hue='Cluster', palette='Set1', s=80, edgecolor='black')

plt.title('K-Means Clusters of Heart Disease Patients (PCA-Reduced)', fontsize=14)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(title='Risk Level')
plt.grid(True)
plt.tight_layout()
plt.show()
```
